You can download PROBA-2/SWAP (https://proba2.sidc.be/about/SWAP) extreme ultraviolet (EUV) images from an archive of SWAP data products (http://proba2.oma.be/swap/data/bsd).

```python
import astropy.io.fits as fits

FITS_path = "./Data/swap_lv1_20120606_000010.fits"

FITS = fits.open(FITS_path)
FITS.info()
```

```python
data1  = FITS[0].data
header1 = FITS[0].header

#----------------- Or ----------------------- 
#data1 = fits.getdata(FITS_path, 0)
#header1 = fits.getheader(FITS_path, 0)
#-------------------------------------------- 

print("Telescope: ", header1["TELESCOP"])
print("Date/Time of the Observation: ", header1["DATE-OBS"])
print("Arcsec per pixel scale :", header1["CDELT1"])
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.figure(figsize=(5,5))
plt.imshow(data1,cmap='gray', origin='lower')
plt.show()
```

```python
import numpy as np

data_array = np.array(data1)
data_array[data_array<1] = 1

log_image = np.log(data_array)
```

```python
plt.figure(figsize=(5,5))
plt.imshow(log_image,cmap='gray', origin='lower')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.imshow(log_image,cmap='hot', origin='lower')
plt.show()
```

```python
cropped_image = log_image[500:800, 300:600]

plt.figure(figsize=(5,5))
plt.imshow(cropped_image,cmap='gray', origin='lower')
plt.show()
```
