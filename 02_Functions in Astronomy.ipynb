Planck's law of Black-body Radiation
=============

```python
import numpy as np
```

--------------------------
Planck's Radiation Formula

```python
def planck(lamb, T):
    
    h = 6.6260*10**(-34)  # Planck's constant [Js]
    k = 1.3806*10**(-23)  # Boltzmann's constant [J/K]
    c = 2.9979*10**(8)    # the velocity of light [m/s]
    
    B = (2*h*c**2)/(lamb**5)*(1/(np.exp((h*c)/(lamb*k*T))-1))
    
    return B
```

-------------
Calculate the Formula and Unit Transfer

```python
um_Wavelengths = np.linspace(0.1, 3, 100)   # 0.1 ~ 3 [um]
K_Temperature = 5000    # [K]
```

```python
um_to_m = 10**(-6)
m_Wavelengths = um_Wavelengths*um_to_m
```

```python
Spectral_Radiance1 = planck(m_Wavelengths, K_Temperature)    # [W/sr/m^-3]
```

```python
nm_to_m = 10**(-9)
kW_to_W = 10**(-3)

Re_Spectral_Radiance1 = Spectral_Radiance1*nm_to_m*kW_to_W    # [kW/sr/m^-2/nm^-1]
```

-------------------
Python Plotting
========


Matplotlib (https://matplotlib.org/) is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(um_Wavelengths, Re_Spectral_Radiance1, label = "T = "+str(K_Temperature)+"K")

plt.legend()

plt.title("Planck's law")
plt.xlabel("Wavelength [$\mu m$]")
plt.ylabel("Spectral radiance [$kW \cdot sr^{-1} \cdot m^{-2} \cdot nm^{-1}$]")

plt.show()
```

-----------------------
Plotting multiple lines in 

```python
um_Wavelengths = np.linspace(0.1, 3, 100)   # 0.1 ~ 3 [um]
m_Wavelengths = um_Wavelengths*um_to_m

K_Temperatures = [3000, 4000, 5000]     # [K]
num_K = len(K_Temperatures)


plt.figure(figsize=(8,5))

for K_Temp1 in K_Temperatures :
    
    Spectral_Radiance1 = planck(m_Wavelengths, K_Temp1)
    Re_Spectral_Radiance1 = Spectral_Radiance1*nm_to_m*kW_to_W
    
    plt.plot(um_Wavelengths, Re_Spectral_Radiance1, label = "T = "+str(K_Temp1)+"K")

plt.legend()

plt.title("Planck's law")
plt.xlabel("Wavelength [$\mu m$]")
plt.ylabel("Spectral radiance [$kW \cdot sr^{-1} \cdot m^{-2} \cdot nm^{-1}$]")

plt.show()
```
